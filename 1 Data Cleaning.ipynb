Create column with total number of campers and flag what parks allow camping based on this

```python
df_parks['Campers'] = df_parks.RVCampers + df_parks.TentCampers + df_parks.BackcountryCampers
```

```python
df_campflag = df_parks.groupby(['ParkName','Year'], as_index = False)['Campers'].sum()
df_campflag['CampFlag'] = ['NoCamping' if x == 0 else 'Camping' for x in df_campflag['Campers']]
df_campflag = df_campflag.drop(columns = ['Campers'])
df_campflag
```

```python
df_parks = df_parks.merge(df_campflag, on = ['ParkName','Year'], how = 'left')
```

Merge Park Locations and NPS Lands datasets to get location of NP and NMs (the most popular units of the NPS)

```python
df_parks.shape
```

```python
df_merged = df_parks.merge(df_location, on = 'ParkName', how = 'left')
```

```python
df_merged.shape
```

Add column to data that is a sum of Recreation Visitors for a given year

```python
#add column with annual visitors
df_sum = df_merged.groupby(['ParkName','Year'], as_index = False)['RecreationVisits'].sum()
df_sum.rename(columns = {'RecreationVisits' : 'AnnualRecVisits'}, inplace = True)
df_sum
```

```python
#combine annual visits tables with main table
df = df_merged.merge(df_sum, on = ['ParkName','Year'], how = 'left')
```

```python
df.head()
```

```python
#create column to show what % of visitors visit in each month
df['MonthlyPctOfAnnualRecVisits'] = round(df.RecreationVisits / df.AnnualRecVisits,4)
df['MonthlyPctOfAnnualRecVisits'] = df['MonthlyPctOfAnnualRecVisits'].fillna(0)
```

In the data some NPs that gained the NP designation post 1979 are labeled as a NP for all years. For example White Sands NP became a NP in 2019 but the data labels it as a NP from 1979 to 2021. The code below changes the park type to National Monument for the years before a park was called a National Park

```python
a1980 = ['Gates of the Arctic',
    'Glacier Bay',
    'Katmai',
    'Kenai Fjords',
    'Kobuk Valley',
    'Lake Clark',
    'Wrangell-St. Elias',
    'Biscayne',
    'Channel Islands']
a1994 = ['Death Valley','Joshua Tree','Saguaro']
```

```python
df.loc[(df['ParkName'].isin(a1980)) & (df['Year']  < 1980), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'].isin(a1994)) & (df['Year']  < 1994), 'ParkType'] = 'National Monument'

df.loc[(df['ParkName'] == 'Great Basin') & (df['Year']  < 1986), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'National Park of American Samoa') & (df['Year']  < 1988), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Dry Tortugas') & (df['Year']  < 1992), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Black Canyon of the Gunnison') & (df['Year']  < 1999), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Cuyahoga Valley') & (df['Year']  < 2000), 'ParkType'] = 'National Recreation Area'
df.loc[(df['ParkName'] == 'Congaree') & (df['Year']  < 2003), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Great Sand Dunes') & (df['Year']  < 2004), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Pinnacles') & (df['Year']  < 2013), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Gateway Arch') & (df['Year']  < 2018), 'ParkType'] = 'National Memorial'
df.loc[(df['ParkName'] == 'White Sands') & (df['Year']  < 2019), 'ParkType'] = 'National Monument'
df.loc[(df['ParkName'] == 'Indiana Dunes') & (df['Year']  < 2019), 'ParkType'] = 'National Lakeshore'
df.loc[(df['ParkName'] == 'New River Gorge') & (df['Year']  < 2020), 'ParkType'] = 'National River'
```
