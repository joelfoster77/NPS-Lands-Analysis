```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import decimal
import os
path = r'C:\Users\joelf\OneDrive\Documents\NPSAnalysis'
```

```python
df = pd.read_csv(os.path.join(path, 'Parks and Location.csv'))
```

```python
df.head()
```

```python
df_corr = df[['RecreationVisits','RecreationHours', 'Month','Campers']]
```

```python
df_corr.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(7,7))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_corr.corr(), annot = True, ax = ax) #annot argument allows annotation
```

There is a stronger coorelation between Campers and RecreationHours than between Campers and RecreationVisits. I would guess this is because campers spend a much larger number of hours per visit than NonCampers

Assuming coorelation between Month and visitors is low because it is polynomial, not linear, as it spikes in summer months and dips in winter.

```python
sns.lmplot(x = 'RecreationVisits', y = 'RecreationHours', data = df_corr)
```

```python
sns.lmplot(x = 'Campers', y = 'RecreationHours', data = df_corr)
```

```python
plots = sns.pairplot(df_corr)
```

The strongest coorelations that are useful in a real world sense are the link between recreationvisits and recreation hours, and the link between campers and recreation hours. I suspect looking at a subset of NPs and NMs would have an even strong corr(campers, recreationhours) as those parks are more likely to allow campers.

```python
sns.regplot(x = 'Month', y='MonthlyPctOfAnnualRecVisits', data = df, order = 2)
```
